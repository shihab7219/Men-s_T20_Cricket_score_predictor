```python
#Important Features Extraction

dataFrame.drop(columns=["venue"],inplace=True)
```

```python
dataFrame['city']
```

```python
eligible_cities = dataFrame['city'].value_counts()[dataFrame['city'].value_counts() > 120].index.tolist()
```

```python
eligible_cities
```

```python
dataFrame = dataFrame[dataFrame["city"].isin(eligible_cities)]
```

```python
dataFrame.city.value_counts()
```

```python
dataFrame.city.unique()
```

```python
dataFrame["current_socre"]= dataFrame.groupby("match_id").cumsum()["runs"]
```

```python
dataFrame['over'] = dataFrame['ball'].apply(lambda x:str(x).split(".")[0])
dataFrame['ball_no'] = dataFrame['ball'].apply(lambda x:str(x).split(".")[1])
dataFrame
```

```python
dataFrame['balls_bowled'] = (dataFrame['over'].astype('int')*6) + dataFrame['ball_no'].astype('int')
dataFrame
```

```python
dataFrame['balls_left'] = 120 - dataFrame['balls_bowled']
dataFrame['balls_left'] = dataFrame['balls_left'].apply(lambda x:0 if x<0 else x)
dataFrame
```

```python
dataFrame['player_dismissed'] = dataFrame['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
dataFrame['player_dismissed'] = dataFrame['player_dismissed'].astype('int')
dataFrame['player_dismissed'] = dataFrame.groupby('match_id').cumsum()['player_dismissed']
dataFrame['wickets_left'] = 10 - dataFrame['player_dismissed']
```

```python
dataFrame['crr'] = (dataFrame['current_socre']*6)/dataFrame['balls_bowled']
```

```python
groups = dataFrame.groupby('match_id')

match_ids = dataFrame['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
dataFrame['last_five'] = last_five
```

```python
dataFrame[['batting_team','bowling_team','city','current_socre','balls_left','wickets_left','crr','last_five']]
```

```python
final_df = dataFrame.groupby('match_id').sum()['runs'].reset_index().merge(dataFrame,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_socre','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
#Separate Dataset

X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#Importing Important libraries for model building

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
#Encoding features

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
#Create Pipline

pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
#Prediction and Evaluation

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
#Save model for deplyment

pickle.dump(pipe,open('pipe2.pkl','wb')) 
```
